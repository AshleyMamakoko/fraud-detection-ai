```python
#Load libraries and dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("../data/synthetic_fraud_dataset.csv")

df.head()
```

```python
#tells the dataset structure
df.info()
df.describe()
```

```python
#Check Class Imbalance (fraud vs non-fraud)
fraud_counts = df["is_fraud"].value_counts()
fraud_counts

# plot it
fraud_counts.plot(kind="bar")
plt.title("Fraud vs Non-Fraud Transaction Count")
plt.xlabel("Class (0 = genuine, 1 = fraud)")
plt.ylabel("Count")
plt.show()
```

```python
#Distribution of Transaction Amounts
plt.figure(figsize=(10,5))
plt.hist(df["amount"], bins=50)
plt.title("Transaction Amount Distribution")
plt.xlabel("Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
#Fraud Amount Comparison
plt.figure(figsize=(10,5))
plt.hist(df[df.is_fraud == 1]["amount"], bins=50, alpha=0.6, label="Fraud")
plt.hist(df[df.is_fraud == 0]["amount"], bins=50, alpha=0.6, label="Not Fraud")
plt.title("Fraud vs Non-Fraud Transaction Amounts")
plt.xlabel("Amount")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
#Fraud by Country
fraud_by_country = df.groupby("country")["is_fraud"].mean().sort_values(ascending=False)

fraud_by_country.plot(kind="bar", figsize=(10,5))
plt.title("Average Fraud Rate by Country")
plt.xlabel("Country")
plt.ylabel("Fraud Rate")
plt.show()
```

```python
#Fraud by Time of Day
df["hour"] = df["time_seconds"] // 3600

fraud_by_hour = df.groupby("hour")["is_fraud"].mean()

fraud_by_hour.plot(figsize=(10,5))
plt.title("Fraud Rate by Hour of the Day")
plt.xlabel("Hour")
plt.ylabel("Fraud Rate")
plt.show()
```
